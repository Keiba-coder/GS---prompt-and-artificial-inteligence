```python
import google.generativeai as genai
from google.colab import userdata
import random
import time
import pandas as pd

# --- 1. CONFIGURAÇÃO DA IA ---
try:
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    genai.configure(api_key=GOOGLE_API_KEY)
    model = genai.GenerativeModel('gemini-1.5-flash')
    ai_disponivel = True
except:
    ai_disponivel = False
    print('Aviso: Chave GOOGLE_API_KEY não configurada nos Secrets.')

# --- 2. FUNÇÕES DO SISTEMA ---

def simular_satelites(n_satelites=3):
    constelacao = []
    for i in range(1, n_satelites + 1):
        sat = {
            'id': f'SAT-{i:03d}',
            'temperatura': random.uniform(10, 100),
            'energia': random.uniform(5, 100),
            'comunicacao': random.choice(['Estável', 'Instável', 'Sem Sinal']),
            'status_modulos': random.uniform(50, 100),
            'alertas': [],
            'acoes': []
        }

        # Lógica de Decisão Automática
        if sat['energia'] < 20:
            sat['alertas'].append('Bateria Crítica')
            sat['acoes'].append('Ativar modo de economia')
        if sat['temperatura'] > 80:
            sat['alertas'].append('Superaquecimento')
            sat['acoes'].append('Reduzir processamento')
        if sat['comunicacao'] != 'Estável':
            sat['alertas'].append('Falha de Comunicação')
            sat['acoes'].append('Redirecionar tráfego')
        if sat['status_modulos'] < 70:
            sat['alertas'].append('Módulos Comprometidos')
            sat['acoes'].append('Solicitar inspeção técnica')

        constelacao.append(sat)
    return constelacao

def formatar_painel(dados):
    df = pd.DataFrame(dados)
    df_fmt = df.copy()
    df_fmt['temperatura'] = df_fmt['temperatura'].apply(lambda x: f"{int(x)}°C")
    df_fmt['energia'] = df_fmt['energia'].apply(lambda x: f"{int(x)}%")
    df_fmt['status_modulos'] = df_fmt['status_modulos'].apply(lambda x: f"{int(x)}%")

    # Renomeando coluna para exibição amigável
    df_fmt = df_fmt.rename(columns={'status_modulos': 'Status dos Módulos'})

    return df_fmt[['id', 'temperatura', 'energia', 'comunicacao', 'Status dos Módulos', 'alertas', 'acoes']]

def gerar_relatorio_final(n=5):
    print("="*60)
    print("SISTEMA DE MONITORAMENTO")
    print("="*60)

    frota = simular_satelites(n)
    display(formatar_painel(frota))

    print("\n--- LOG DE ALERTAS ---")
    tem_alerta = False
    for s in frota:
        if s['alertas']:
            tem_alerta = True
            print(f"[!] {s['id']}: {', '.join(s['alertas'])} | Ação: {', '.join(s['acoes'])}")
    if not tem_alerta: print("[*] Todos os sistemas operando normalmente.")

    if ai_disponivel:
        print("\n--- DIAGNÓSTICO ESTRATÉGICO (GEMINI AI) ---")
        prompt = f"Aja como um engenheiro aeroespacial. Analise esta telemetria e dê um resumo executivo: {frota}"
        try:
            res = model.generate_content(prompt)
            print(res.text)
        except:
            print("Erro na chamada da API.")

# --- 3. EXECUÇÃO ---
gerar_relatorio_final(5)
```
